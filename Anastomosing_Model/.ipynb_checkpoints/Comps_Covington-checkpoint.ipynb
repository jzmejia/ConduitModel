```python
from scipy.integrate import odeint
```

```python
# Constants
alpha = 5./4.
rho_w = 1000.#density of water, kg/m^3
rho_i = 910#density of ice, kg/m^3
L_f = 3.35e5#Latent heat of fusion (ice), J/kg
n=3.#ice flow law exponent, unitless
f=0.0375#Darcy-Weisbach friction factor, unitless
g = 9.81#gravitational accel, m^2/s
A = 6E-24
```

```python
#initial values
AR = 15
L = 100
Zi = 700
h0 = 100
S0 = 0.1
#constants of integration
C1 = 1/(rho_i*L_f)
C2 = 2*A*n**-n
C3 = 2**.25*sqrt(pi+2)/(pi**.25*sqrt(rho_w*f))
Pi = rho_i*g*Zi
```

```python

def dy_dt(y,t,Ri,Rf):
    if t > sw:
        R = Rf
    else:
        R = Ri
    h1 = y[0]
    S1 = y[1]    
    dh_dt = (R-C3*(S1**1.25)*sqrt((rho_w*g*h1)/L))/AR
    dS_dt = C1*C3*(S1**alpha)*((rho_w*g*h1)/L)**1.5 - C2*S1*(Pi-(rho_w*g*h1))**3
    return [dh_dt,dS_dt]


# Run model with above specification
nhours = 10
nsecs = nhours*60*60
t = linspace(0,nsecs,500)
sw = t[250]

# integrate for var Rf/Ri
sol125 = odeint(dy_dt, [h0,S0], t,args=(1,1.25))
sol1 = odeint(dy_dt, [h0,S0], t,args=(1,1.5))
sol2 = odeint(dy_dt, [h0,S0], t,args=(1,2))
sol25 = odeint(dy_dt, [h0,S0], t,args=(1,2.5))
sol3 = odeint(dy_dt, [h0,S0], t,args=(1,3))
solp5 = odeint(dy_dt, [h0,S0], t,args=(0.5,1.5))
solp75 = odeint(dy_dt, [h0,S0], t,args=(0.75,1.5))
# plot time in minutes
t_min_plot = linspace(0,nhours*60,500)
```

```python
# plot head
figure()
plot(t_min_plot,sol125[:,0])
plot(t_min_plot,sol1[:,0])
plot(t_min_plot,sol2[:,0])
plot(t_min_plot,sol25[:,0])
plot(t_min_plot,sol3[:,0])
xlabel('Time (min)')
ylabel('Hydraulic Head in Moulin ($m$)')
legend(['$R_f$=1.25m','$R_f$=1.5m','$R_f$=2m','$R_f$=2.5m','$R_f$=3m'])
savefig('head.png')
# plot S
figure()
plot(t_min_plot,sol125[:,1])
plot(t_min_plot,sol1[:,1])
plot(t_min_plot,sol2[:,1])
plot(t_min_plot,sol25[:,1])
plot(t_min_plot,sol3[:,1])
xlabel('Time (min)')
ylabel('Cross-sectional Area ($m^2$)')
legend(['$R_f$=1.25m','$R_f$=1.5m','$R_f$=2m','$R_f$=2.5m','$R_f$=3m'],loc='upper left')
savefig('S.png')

```

```python
#find time to equlize after Rf
numblist=[]
difflist=[]
a=1
for i, val in enumerate(sol125[250:,1]):
    indx = 250+i
    dif = abs(val - a)
    print i, indx, dif 
    difflist.append(dif)
    if dif < 0.00001:
        numblist.append(i)
    a = val
```
